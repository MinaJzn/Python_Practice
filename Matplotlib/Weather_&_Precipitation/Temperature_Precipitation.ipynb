# Average Temperature and Average Precipitation in Japan (2024)
Line plot shows **Average Temperature(Â°C)** and bar plot shows **Average Precipitation(mm)** for each month in Japan during 2024.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file=r'D:\Data Visualization\Matplotlib\Practice\Weather\Temperature_Precipitation.csv'
dataframe=pd.read_csv(file)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,4))

ax1.plot(dataframe['Month'],dataframe['AvgTemp_C'],color='hotpink')
ax1.set(
    xlabel='Month',
    ylabel='Average Temperature(Â°C)',
    yticks=np.arange(0, 36 , 3),
)
ax1.grid(True,alpha=.35)
for i , temp in enumerate(dataframe['AvgTemp_C']):
    ax1.text(i , temp ,str(temp),ha='center')


ax2.bar(dataframe['Month'],dataframe['Precipitation_mm'],color='#48D1CC')
ax2.set(
    xlabel='Month',
    ylabel='Average Precipitation(mm)',
)
ax2.grid(True,alpha=.35)
for i, v in enumerate(dataframe['Precipitation_mm']):
    ax2.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```
