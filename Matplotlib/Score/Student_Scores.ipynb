# Exam Score Distribution
This plot shows the distribution of exam score in class.
- Histogram: The count (or density) of students within score ranges
- KDE curve: The estimated density to visualize the overall pattern

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
file=r'exam_scores.csv'
dataframe=pd.read_csv(file)
```

```python
fig,ax =plt.subplots(figsize=(15,5))
sns.histplot(dataframe['score'],kde=True,bins=30,color='#DC143C',edgecolor='black',alpha=0.4)
ax.set(
    xlabel='Score',
    ylabel='Density',
    yticks=np.arange(0, 24 , 2)
)

plt.title('Exam Score Histogram with KDE')
plt.show()
```
