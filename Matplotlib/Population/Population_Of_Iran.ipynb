# Population Distribution by Age and Gender in Iran (2024)

This code reads population data from a CSV file and creates a grouped bar chart showing the population of males and females across different age groups in Iran for the year 2024. The y-axis labels are formatted to display values in millions for better readability.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
```

```python
file=r'D:\Data Visualization\Matplotlib\Practice\Population\Iran-2024.csv'
dataframe=pd.read_csv(file)
```

```python
fig,ax=plt.subplots(figsize=(12,6))

x=np.arange(len(dataframe['Age']))
width=.35

plt.bar(x-width/2,dataframe['M'],width,label="Male")
plt.bar(x+width/2,dataframe['F'],width,label="Female",color='pink')

plt.xticks(x,dataframe['Age'])
plt.xlabel('Age Group')
plt.ylabel('Population')
plt.legend(title='Gender:')


plt.yticks(np.arange(0, 5000001, 500000))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x*1e-6:.1f}M'))

plt.show()
```
