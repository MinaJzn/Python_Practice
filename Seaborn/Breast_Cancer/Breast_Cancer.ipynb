# Data Visualization on the Breast Cancer Dataset:
_A simple project to practice data visualization techniques on the Breast Cancer dataset_

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
x,y=load_breast_cancer(as_frame=True,return_X_y=True)
```

```python
x
```

```python
df=pd.concat([x,y],axis=1)
df
```

```python
df.info()
```

```python
df.columns
```

### Column Descriptions – Breast Cancer Dataset

#### Mean Features
- **mean radius** – Mean radius of the tumor (distance from center to edge).
- **mean texture** – Mean texture, measuring pixel intensity variations.
- **mean perimeter** – Mean perimeter of the tumor.
- **mean area** – Mean area of the tumor.
- **mean smoothness** – Mean smoothness of tumor edges.
- **mean compactness** – Mean compactness (area vs. perimeter²).
- **mean concavity** – Mean concavity (severity of concave parts of the contour).
- **mean concave points** – Mean number of concave points on the contour.
- **mean symmetry** – Mean symmetry of the tumor.
- **mean fractal dimension** – Mean fractal dimension (complexity of the contour).

#### Error Features (Standard Error)
- **radius error** – Standard error of the radius.
- **texture error** – Standard error of the texture.
- **perimeter error** – Standard error of the perimeter.
- **area error** – Standard error of the area.
- **smoothness error** – Standard error of edge smoothness.
- **compactness error** – Standard error of compactness.
- **concavity error** – Standard error of concavity.
- **concave points error** – Standard error of concave points.
- **symmetry error** – Standard error of symmetry.
- **fractal dimension error** – Standard error of fractal dimension.

#### Worst Case Features
- **worst radius** – Worst (largest) radius of the tumor.
- **worst texture** – Worst (highest) texture value.
- **worst perimeter** – Worst perimeter value.
- **worst area** – Worst area of the tumor.
- **worst smoothness** – Worst edge smoothness.
- **worst compactness** – Worst compactness value.
- **worst concavity** – Worst concavity value.
- **worst concave points** – Worst number of concave points.
- **worst symmetry** – Worst symmetry value.
- **worst fractal dimension** – Worst fractal dimension.

#### Target
- **target** – Target label: 1 = malignant, 0 = benign.



❗ Based on the heatmap and the analysis of the absolute correlation values, and due to the high correlation between the perimeter and area features with mean radius, radius error, and worst radius, plotting separate charts for these features is redundant. Therefore, analyzing the radius feature alone provides sufficient insight.

```python
sns.heatmap(
    abs(df.filter(regex='mean|target').corr()),
    annot=True,
    cmap='Reds'
)
```

```python
sns.heatmap(
    abs(df.filter(regex='error|target').corr()),
    annot=True,
    cmap='Reds'
)
```

```python
sns.heatmap(
    abs(df.filter(regex='worst|target').corr()),
    annot=True,
    cmap='Reds'
)
```

```python
df.drop(columns=['mean perimeter', 'mean area','perimeter error', 'area error','worst perimeter', 'worst area'],inplace=True)
```

```python
df_mean=df.filter(regex='mean|target')
df_mean
```

```python
sns.displot(
    data=df_mean,
    x='mean radius',
    hue='target'
)
```

```python
sns.displot(
    data=df_mean,
    x='mean texture',
    hue='target'
)
```

As observed below, a negative correlation with the target variable indicates an inverse relationship. For instance, a decrease in the mean radius is associated with an increased probability of the tumor being malignant.

```python
sns.heatmap(
    df_mean.corr(),
    annot=True,
    cmap='coolwarm'
)
```

```python
sns.displot(
    data=df_mean,
    x='mean symmetry',
    hue='target'
)
```

```python
sns.displot(
    data=df_mean,
    x='mean smoothness',
    hue='target'
)
```

```python
sns.regplot(
    data=df_mean,
    x='mean radius',
    y='target',
    logistic=True
)
```

As observed below, the feature `mean fractal dimension` does not have a significant correlation with the `target` variable. Therefore, to optimize resources and reduce experimental costs in future analyses, it is recommended to exclude this feature.

```python
sns.regplot(
    data=df_mean,
    x='mean fractal dimension',
    y='target',
    logistic=True
)
```

According to the heatmap below, the `error features` do not show significant correlation with the `target` variable, and most likely, removing them from the feature set will not have a considerable impact on the analysis results.

```python
sns.heatmap(
    df.filter(regex='error|target').corr(),
    annot=True,
    cmap='Reds'
)
```

```python
sns.heatmap(
    df.filter(regex='worst|target').corr(),
    annot=True,
    cmap='Reds'
)
```

```python
sns.displot(
    data=df.filter(regex='worst|target'),
    x='worst radius',
    hue='target'
)
```
