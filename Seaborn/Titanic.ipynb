### Correlation Analysis:
Based on the correlation heatmap (absolute values), only `fare`, `pclass`, and `who` show a noticeable correlation with the `survived` feature.
Among them, `who` has the strongest correlation (0.47), followed by `pclass` (0.34) and `fare` (0.26).

```python
df['alone']=df['alone'].map({
    True:1,
    False:0
})

df['who']=df['who'].map({
    'man':1,
    'woman':2,
    'child':3
})
df_numerical=df.select_dtypes(include=['int64','float64'])
```

```python
sns.heatmap(
    abs(df_numerical.corr()),
    annot=True,
    cmap='Reds'
)
```
